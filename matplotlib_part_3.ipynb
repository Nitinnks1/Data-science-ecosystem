# **python matplotlib part 3**
# **ploting bar chart**

```python
import matplotlib.pyplot as plt
import numpy as np
import random
from matplotlib import style
```

```python
#dataset of indian artificial intelligence production class
# data set of class [python, R, A.I, Data Science, Machine learning]

classes = ["python", "R", "A.I", "DS", "ML"]

class1 = [20,25,15,10,30]
class2 = [25,15,30,20,30]
class3 = [20,30,12,35,15]
```

```python
plt.bar(classes, class1 )
```

```python
plt.barh(classes, class1 )
```

```python
plt.bar(classes, class1, width = 0.5, align = 'edge', color= 'c'  )
```

```python
plt.bar(classes, class1, width = 0.5, align = 'edge', color= 'c', edgecolor = 'm', linewidth = '4', linestyle = '--',
        visible = True)
plt.title('class1 student')
```

```python

classes_index = np.arange(len(classes))
width = 0.3

plt.bar(classes_index+width, class1, width, color= 'c')
plt.bar(classes_index, class2, width, color= 'r')


plt.title('class1 & class2 student')
plt.xlabel('student course')
plt.ylabel('no. of student')
plt.xtricks =(classes_index+width, classes)
```

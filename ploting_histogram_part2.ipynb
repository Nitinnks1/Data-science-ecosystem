  # **Ploting Histogram**



```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
boy_voter = np.random.randint(18,50,(100))
girl_voter = np.random.randint(18,50,(100))
```

```python

print(boy_voter)
print(girl_voter)
```

```python
plt.hist(boy_voter)
plt.xlabel('voter age')
plt.ylabel('voter ratio')
plt.title('boy voter')
plt.show()

```

```python
bins = [20,25,30,35,40,45,50,55,60]
plt.hist(boy_voter,bins,  rwidth =0.8, color = 'c', histtype = 'bar', orientation='horizontal')
plt.xlabel('voter age')
plt.ylabel('voter ratio')
plt.title('boy voter')
plt.show()
```

```python
bins = [20,25,30,35,40,45,50]
plt.figure(figsize=(10,6))
plt.hist([boy_voter, girl_voter],bins,  rwidth =0.8, color = ['c','m'], histtype = 'bar', orientation='vertical', label = ['boy voter','girl voter'])


plt.legend()



plt.xlabel('voter age')
plt.ylabel('voter ratio')
plt.title('boy & girl voter chart ')
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")
bins = [20,25,30,35,40,45,50]
plt.figure(figsize=(10,6))
plt.hist([boy_voter, girl_voter],bins,  rwidth =0.8, color = ['c','m'], histtype = 'bar', orientation='vertical', label = ['boy voter','girl voter'])


plt.legend()



plt.xlabel('voter age')
plt.ylabel('voter ratio')
plt.title('boy & girl voter chart ')
plt.show()
```
